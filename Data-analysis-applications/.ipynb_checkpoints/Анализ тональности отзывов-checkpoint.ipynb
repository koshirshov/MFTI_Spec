```python
from nltk.corpus import movie_reviews
negits = movie_reviews.fileids('neg')
posits = movie_reviews.fileids('pos')
```

```python
negits[:5]
```

```python
movie_reviews.words(fileids=negits[0])
```

```python
negfeats = [' '.join(movie_reviews.words(fileids=[f])) for f in negits]
posfeats = [' '.join(movie_reviews.words(fileids=[f])) for f in posits]
texts = negfeats + posfeats
labels = [0] * len(negfeats) + [1] * len(posfeats)
```

```python
negfeats[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer, TfidfVectorizer
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.svm import LinearSVC
from sklearn.cross_validation import cross_val_score
from sklearn.pipeline import Pipeline
```

```python
def text_classifier(vectorizer, transformer, classifier):
    return Pipeline(
    [('vectorizer', vectorizer),
     ('transformer', transformer),
     ('classifier', classifier)]
    )
```

```python
pipeline1 = Pipeline(steps=[('vectorizer', CountVectorizer())
          ,('transformer', TfidfTransformer())
          ,('classifier', LinearSVC())])
```

```python
cross_val_score(pipeline1, texts, labels)
```

```python
for clf in [LogisticRegression, LinearSVC, SGDClassifier]:
    print(clf)
    print(cross_val_score(text_classifier(CountVectorizer(), TfidfTransformer(), clf()), texts, labels).mean())
```

```python
cs = CountVectorizer()
cs.fit(texts)
bbb = cs.transform(texts)
cs.get_feature_names()
```

```python
cs1 = TfidfTransformer()
cs1.fit(bbb)
print(cs1.transform(bbb)[0,:])
```

```python
#классификатор на всех текстах
clf_pipeline = Pipeline(
[ ('vectorizer', TfidfVectorizer()),
('classifier', LinearSVC()) ]
)
clf_pipeline.fit(texts, labels)
```

```python
clf_pipeline.predict(['amazing film!'])
```

```python
from sklearn.decomposition import NMF, TruncatedSVD

v = CountVectorizer()
mx = v.fit_transform(texts)
mf = NMF(10)
u = mf.fit_transform(mx)
```

```python
%%time
#for transform in [TruncatedSVD, NMF]:
for transform in [TruncatedSVD]:
    print(transform)
    print(cross_val_score(text_classifier(CountVectorizer()
                                    , transform(n_components = 1000)
                                    , LinearSVC())
                                    , texts, labels).mean()
         )
    print('\n')
    #084 вродще получается, долго на 1000
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
cross_val_score(
    Pipeline(
    [
        ('vectorizer', CountVectorizer()),
        ('transformer', TruncatedSVD(100)),
        ('classifier', RandomForestClassifier(100))
    ]
    ),
    texts, labels
    )
```

```python
#увеличим число призн и деревьев, лучше не стало(хуже)
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
cross_val_score(
    Pipeline(
    [
        ('vectorizer', CountVectorizer()),
        ('transformer', TruncatedSVD(1000)),
        ('classifier', RandomForestClassifier(1000))
    ]
    ),
    texts, labels
    )
```

```python
#Tfidf
from sklearn.ensemble  import RandomForestClassifier, GradientBoostingClassifier
cross_val_score(
    Pipeline(
    [
        ('vectorizer', TfidfVectorizer()),
        ('transformer', TruncatedSVD(1000)),
        ('classifier', RandomForestClassifier(1000))
    ]
    ),
    texts, labels
    )
```

```python
#cовмещаем тфидф и svd
from sklearn.pipeline import FeatureUnion
estimators = [('tfidf', TfidfTransformer()), ('svd', TruncatedSVD(1) )]
combined = FeatureUnion(estimators)
```

```python
cross_val_score(
    Pipeline(
    [
        ('vectorizer', CountVectorizer()),
        ('transformer', combined),
        ('classifier', LinearSVC())
    ]
    ),
    texts, labels
    )
```
