```python
from matplotlib.colors import ListedColormap
colors = ListedColormap(['red','yellow'])
def plot_2d_dataset(data,colors):
    pyplot.figure(figsize(8,8))
    pyplot.scatter(list(map(lambda x: x[0], data[0]))
                   ,list(map(lambda x: x[1], data[0]))
                   , c = data[1], cmap = colors)
plot_2d_dataset(circles_noise, colors)
```

```python
import pandas as pd
iris = datasets.load_iris()
iris.feature_names
iris.target_names
iris_df = pd.DataFrame(iris.data)
iris_df.columns = iris.feature_names
iris_df['target'] = iris.target
iris_df['target_names'] = iris_df['target'].apply(lambda x: iris.target_names[x]) 
iris_df.head()
```

```python
#4 строки, 3 cтолбца, 
#subplot - матрица из графиков
pyplot.figure(figsize(20,24))
plot_number = 0
for feature_name in iris_df.iloc[:,0:4].columns:
    for target_name in iris_df['target_names'].unique():
        plot_number += 1
        pyplot.subplot(4,3,plot_number)
        pyplot.hist(iris_df[iris_df['target_names'] == target_name][feature_name])
        pyplot.title(target_name)
        pyplot.ylabel(feature_name)
        pyplot.xlabel('cm')
```

```python
blobs = datasets.make_blobs(centers = 2, cluster_std = 5.5, random_state =1)
X_train, X_test, y_train, y_test = cross_validation.train_test_split(
    blobs[0], blobs[1], test_size = 0.3, random_state = 1)
ridge_classifier = linear_model.RidgeClassifier(random_state = 1)
ridge_classifier.fit(X_train, y_train)
ridge_predictions = ridge_classifier.predict(X_test)
print(y_test)
print(ridge_predictions)
```


# Cтратегии проведения кросс вал

```python
#Kfold
for train_indices, test_indices in cross_validation.KFold(10, n_folds = 5):
    print(train_indices, test_indices)
```

```python
#Kfold
for train_indices, test_indices in cross_validation.KFold(10
                                        , n_folds = 5,shuffle=True):
    print(train_indices, test_indices)
```

```python
#StratifiedKFold
#cохраняем соотношение классов в обучающей и тестовой
target = np.array([0] * 5 + [1] * 5)
print(target)
for train_indices, test_indices in cross_validation.StratifiedKFold(target
                                        , n_folds = 2,shuffle=True):
    print(train_indices, test_indices)
#поровну объектов каждоо класса
```

```python
#Shaffle split, c возвращение/ Cлучайное
#получить много выборок, нет огр сколько раз каждый обьект может появиться на обуч или тесте
for train_indices, test_indices in cross_validation.ShuffleSplit(10
                                        , n_iter = 10,test_size = 0.2):
    print(train_indices, test_indices)
```

```python
#Stratify Shaffle split#cбаланс выборки
target = np.array([0] * 5 + [1] * 5)
print(target)
for train_indices, test_indices in cross_validation.StratifiedShuffleSplit(target
                                        , n_iter = 4,test_size=0.2):
    print(train_indices, test_indices)
```

```python
#Leave-One-Out
for train_indices, test_indices in cross_validation.LeaveOneOut(10):
    print(train_indices, test_indices)
```

# GridSearch

```python
train_data, test_data, train_labels, test_labels= cross_validation.train_test_split(iris.data, iris.target, test_size = 0.3)
classifier = linear_model.SGDClassifier()
from sklearn import cross_validation, grid_search, metrics
```

```python
#Генерация сетки
classifier.get_params().keys()
```

```python
parameters_grid = {
    'loss' : ['hinge', 'log', 'squared_hinge', 'squared_loss'],
    'penalty' : ['l1', 'l2'],
    'n_iter' : list(range(5,10)),
    'alpha' : np.linspace(0.0001, 0.001, num = 5),
    }
```

```python
cv = cross_validation.StratifiedShuffleSplit(train_labels, n_iter = 10
                                             , test_size = 0.2, random_state = 0)
cv
```

```python
grid_cv = grid_search.GridSearchCV(classifier, parameters_grid
                                   , scoring = 'accuracy', cv = cv)
```

```python
%%time
grid_cv.fit(train_data, train_labels)
```

```python
grid_cv.best_estimator_
```

```python
grid_cv.best_score_
grid_cv.best_params_
```
