```python
from sklearn.datasets import load_breast_cancer, load_digits
import numpy as np; import pandas as pd
X_bc = load_breast_cancer().data
y_bc = load_breast_cancer().target
X_d = load_digits().data
y_d = load_digits().target
from sklearn.model_selection import cross_val_score 
from matplotlib import pyplot as plt
%matplotlib inline
from sklearn.naive_bayes import BaseNB, BernoulliNB, MultinomialNB, GaussianNB
```

```python
pd.DataFrame(X_d).head()
```

```python
BernoulliNB1 = BernoulliNB()
MultinomialNB1 = MultinomialNB()
GaussianNB1 = GaussianNB()
```

```python

print(cross_val_score(BernoulliNB1, X = X_bc, y = y_bc, cv = 5))
print(cross_val_score(MultinomialNB1, X = X_bc, y = y_bc, cv = 5))
print(cross_val_score(GaussianNB1, X = X_bc, y = y_bc, cv = 5), 'maxw')
```

```python
print(cross_val_score(BernoulliNB1, X = X_d, y = y_d, cv = 5))
print(cross_val_score(MultinomialNB1, X = X_d, y = y_d, cv = 5), 'maxw')
print(cross_val_score(GaussianNB1, X = X_d, y = y_d, cv = 5))
```

ожидали, что будет качество макс при определенном выборе распределения, тк для дискретных разреженных лучше выбирать мультинорманое, для непрерывных с мелким разбросом - нормальное, для непр с выбросами - более размазанные распре
